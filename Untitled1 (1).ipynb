In this jupyter notebook I am validating the mass sheets input to a CMB lensing map. These mass sheets are:

- z<3: the same lens plane mass maps used to create the galaxy lensing maps. These are created with a modified DTFE method from downsampled particle lightcones in cosmo-DC2. 
- z>3: log-normal realisations of mass sheets, provided by Joachim Harnois-Deraps.

For now I combine these in a simple Born approximation, scaled with the lensing kernel and added without ray-tracing. This is provided in separate z<3 and z>3 maps so that we can properly check scaling consistency and swap out the higher redshift mass sheets if needed. Once these maps are properly validated we can apply ray-tracing. 

```python
#dependencies 
import numpy as np
import healpy as hp
import matplotlib.pyplot as plt
import pylab as plt
import pyccl as ccl
```

Read in the high and low redshift CMB lensing maps and plot them

```python
map_lowz = hp.read_map('/global/projecta/projectdirs/lsst/groups/CS/cosmoDC2/mass_sheets_behind_z3/cmb_maps/cmblensmap_1_masked.fits')
map_hiz = hp.read_map('/global/projecta/projectdirs/lsst/groups/CS/cosmoDC2/mass_sheets_behind_z3/cmb_maps/cmblensmap_2_masked.fits')
hp.mollview(map_lowz+map_hiz)
```

Compute the angular power spectra of the maps (individual and combined)

```python
alm_cmb = hp.map2alm(map_lowz+map_hiz,lmax=2000)
cl_cmb = hp.alm2cl(alm_cmb)*8. # correcting for mask

alm_cmb1 = hp.map2alm(map_lowz,lmax=2000)
cl_cmb1 = hp.alm2cl(alm_cmb1)*8.

alm_cmb2 = hp.map2alm(map_hiz,lmax=2000)
cl_cmb2 = hp.alm2cl(alm_cmb2)*8.
```

Plot the results

```python
plt.loglog(cl_cmb1) # low redshift
plt.loglog(cl_cmb2) # high redshift
plt.xlim([100,2000])
```

Compute theory from CCL

```python
cosmo = ccl.Cosmology(Omega_c=0.22, Omega_b=0.045, h=0.71, A_s=2.168e-9, n_s=0.96)
cmbl = ccl.CMBLensingTracer(cosmo, 1090.)
ell = np.arange(2, 2000)
cls_cmb = ccl.angular_cl(cosmo, cmbl, cmbl, ell)
```

Plot the results

```python
plt.plot((cl_cmb))
plt.plot(ell, cls_cmb, 'k-') # factor of 2 out here for some reason
plt.plot(ell, cls_cmb*2., 'k--') # factor of 2 out here for some reason
plt.xscale('log')
plt.yscale('log')
plt.xlim([100,2000])
plt.xlabel(r'$l$', fontsize=20)
plt.ylabel(r'$C_l$', fontsize=20)
plt.show()
```

```python
zlo = 1.0; zhi= 1.1;
zmask = (cosmodc2_data['redshift_true']<zhi)&(cosmodc2_data['redshift_true']>zlo)&(cosmodc2_data['mag_r']<mag_cut)
kappa = cosmodc2_data['convergence'][zmask]
ra = cosmodc2_data['ra_true'][zmask]
dec = cosmodc2_data['dec_true'][zmask]
print(len(kappa))
```

Create a map of convergence values from the catalog

```python
pixnum = hp.ang2pix(2048,ra,dec,lonlat=True)
print(len(pixnum))
pixvals,counts = np.unique(pixnum, return_counts=True)
weightmap = np.zeros(hp.nside2npix(2048))
weightmap[pixvals] = counts
kappamap = np.zeros(hp.nside2npix(2048))
for i in range(len(pixnum)):
    kappamap[pixnum[i]]+=kappa[i]
for i in range(len(weightmap)):
    if weightmap[i]>0:
        kappamap[i] = kappamap[i]/weightmap[i]
```
